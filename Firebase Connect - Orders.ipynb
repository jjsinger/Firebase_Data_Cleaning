### Create Master DataFrame

```python
# Instantiate Blank Data Frame
# Store Columns in variable
master_df = pd.DataFrame()
cols = orders_dict_only_orders.columns

for single_account in cols:
    #print(orders_dict_only_orders[single_account])
    
    # Flatten
    df = pd.json_normalize(orders_dict_only_orders[single_account])
    
    # Add Account column and fill the value
    df['account'] = single_account
    
    #Concat dataframes
    frames = [master_df, df]
    master_df = pd.concat(frames)
    
    
# master_df
```
